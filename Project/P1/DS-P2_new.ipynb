# Class Creation

```python
class MyRegressor:
    def fit(self,X,y):
        pass

    def predict(self,X):
        pass
    
    def get_root_mean_square_error(self,predicted_values, actual_values):
        return np.sqrt(np.power(predicted_values-actual_values,2).mean())
    
    def get_r2_score(self,predicted_values, actual_values):
        return 1-((predicted_values-actual_values)**2).sum()/((actual_values-actual_values.mean())**2).sum()
    
    def plot_line(self,predicted_values, actual_values):
        plt.scatter(actual_values, predicted_values, color='blue')
        t=np.arange(actual_values.min(),actual_values.max()+1)
        plt.plot(t,t, color='red')
        plt.show()
```

```python
class MyRandomForestRegressor(MyRegressor):
    def __init__(self) -> None:
        self.model = RandomForestRegressor()

    def fit(self, X, y):
        self.model.fit(X, y)

    def predict(self, X):
        return self.model.predict(X)
```

```python
class MySVMRegressor(MyRegressor):
    def __init__(self) -> None:
        self.model = SVR()

    def fit(self, X, y):
        self.model.fit(X, y)

    def predict(self, X):
        return self.model.predict(X)
```

```python
cols = ["Sim Card Type", "DAR", "Log Price No Discount"]
```

```python
def outlier_clean(df, column_list):
    for i in column_list:
        q1 = np.percentile(df[i], 25)
        q3 = np.percentile(df[i], 75)
        IQR = q3 - q1
        lower_bound = q1 - 1.5 * IQR
        upper_bound = q3 + 1.5 * IQR
        mask = (df[i] >= lower_bound) & (df[i] <= upper_bound)
        df = df[mask]
    return df
```

```python
df=pd.read_csv('phones.csv')
df=df.drop(['Unnamed: 0'],axis=1)
le = LabelEncoder()
df['model'] = le.fit_transform(df['Model'])
df['seller'] = le.fit_transform(df['Seller'])
print(df.shape)
df = outlier_clean(df, ['Price No Discount'])
print(df.shape)
```

```python
X = df.drop(cols, axis=1)
X = X[X.select_dtypes(include=np.number).columns]
y = df['Price No Discount']

X_train, X_test, _, _ = train_test_split(X, y, test_size=0.2, random_state=8)
```

```python
X_trains = {}
y_trains = {}
fit_models = {}
unique_models = X_train['model'].unique()
for model in  unique_models:
    X_trains[model] = X_train[X_train["model"] == model]
    y_trains[model] = X_trains[model]["Price No Discount"]
    X_trains[model] = X_trains[model].drop(["Price No Discount", "model", "Brand Num"], axis=1)
    fit_models[model] = RandomForestRegressor()
    fit_models[model].fit(X_trains[model], y_trains[model])
```

```python
X_tests = {}
y_tests = {}
y_preds = {}
all_y_test = []
all_y_pred = []
unique_models = X_test['model'].unique()
for model in  unique_models:
    X_tests[model] = X_test[X_test["model"] == model]
    y_tests[model] = X_tests[model]["Price No Discount"]
    X_tests[model] = X_tests[model].drop(["Price No Discount", "model", "Brand Num"], axis=1)
    y_preds[model] = fit_models[model].predict(X_tests[model])
    print(root_mean_squared_error(y_tests[model], y_preds[model]))
    all_y_test.extend(y_tests[model])
    all_y_pred.extend(y_preds[model])
```
